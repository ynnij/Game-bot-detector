```python
import pandas as pd
dataset1=pd.read_csv('Player information features.csv')
dataset2=dataset1.replace('Human',0)
dataset=dataset2.replace('Bot',1)

x=dataset.fillna(value=0) 
x=x.astype('float32') 

dataset.pop('Actor')
dataset.pop('A_Acc')

from sklearn.utils import shuffle 
x=shuffle(x)

from sklearn.metrics import confusion_matrix 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()

X_len=len(x.columns)-1 
y_len=len(x.columns)-1 
X=x.iloc[:,0:X_len] 
y=x.iloc[:,y_len]



from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X, y,test_size=0.3,random_state=0)
from sklearn.tree import DecisionTreeClassifier
tree_model=DecisionTreeClassifier(criterion="entropy", max_depth=4, min_samples_split=50)
tree_model.fit(X_train,Y_train)

from sklearn.model_selection import cross_val_score
print('-----DecisionTree-----')
score=cross_val_score(tree_model, X_train, Y_train, cv=5)
print('cross validation score : {:.3f}'.format(score.mean()))

import sklearn.metrics as mt
pred=tree_model.predict(X_test)
accuracy = mt.accuracy_score(Y_test, pred)
recall = mt.recall_score(Y_test, pred)
precision = mt.precision_score(Y_test, pred)
f1_score = mt.f1_score(Y_test, pred)

print("Accuracy: ", format(accuracy))
print('Recall: ', format(recall))
print('Precision: ', format(precision))
print('F1_score: ', format(f1_score))


from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion='entropy',n_estimators=100,n_jobs=2,random_state=123456) 
classifier.fit(X_train, Y_train)
pred_r=classifier.predict(X_test)


print('-----RandomForest-----')

accuracy = mt.accuracy_score(Y_test, pred_r)
recall = mt.recall_score(Y_test, pred_r)
precision = mt.precision_score(Y_test, pred_r)
f1_score = mt.f1_score(Y_test, pred_r)

print("Accuracy: ", format(accuracy))
print('Recall: ', format(recall))
print('Precision: ', format(precision))
print('F1_score: ', format(f1_score))


import numpy as np
from collections import Counter
pred=pred.astype('int64')
print(' DT 결과 ')
np.set_printoptions(threshold=np.inf,linewidth=np.inf)
print(pred.T)
count = Counter(pred.T)  
print("Human 개수:",count[0])
print("Bot 개수:", count[1])



pred_r=pred_r.astype('int64')
print(' RT 결과 ')
np.set_printoptions (threshold=np.inf,linewidth =np.inf)
print(pred_r.T )
count=Counter(pred_r.T)  
print("Human 개수:", count[0])
print("Bot 개수:", count[1])  
```
